### **Найдите самую эффективную платформу по средним значениям конверсий**

Cохраните результат в переменную `most_effective_platform`

````python
import csv
from datetime import datetime
from statistics import mean

lines = []
with open('campaign_data.csv','r', newline='') as f:
  lines.extend(csv.reader(f, delimiter=','))

data = [{'id': int(line[0]),
         'start_date': datetime.strptime(line[1], '%Y-%m-%d'),
         'end_date': datetime.strptime(line[2], "%Y-%m-%d"),
         'type' : line[3],
         'platform' : line[4],
         'budget' : int(line[5]),
         'City' : line[6],
         'Impressions': int(line[7].replace('.0', '')),
         'Clicks' : int(line[8].replace('.0', '')),
         'Conversion' : int(line[9].replace('.0', '')),
         'Revenue' : int(line[10].replace('.0', ''))
         } for line in lines[1:]]

l = len(data)
#platforms = set(all_pf for i in range(l)]) # Не сохраняет порядок

# Создание списка уникальных 'platform' с сохранением порядка
all_pf = [data[i]['platform'] for i in range(l)]
#platforms = list(dict.fromkeys(all_pf for i in range(l)]))    # 1 вариант
_ = set()
platforms = [x for x in all_pf if not (x in _ or _.add(x))]    # 2 вариант
avr_PfCon = {round(mean(data[i]['Conversion'] for i in range(l) if data[i]['platform'] == pform), 2): pform for pform in platforms}
most_effective_platform = avr_PfCon[max(avr_PfCon.keys())]

print(avr_PfCon)
most_effective_platform
````

````python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
# Открываем файл с правильными ответами
with open('marketing_campaigns.txt', 'r') as f:
    answers = f.read().split('\n')

correct_answer = answers[0]

try:
    assert most_effective_platform == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
````

### **Напишите функцию, которая возвращает средний бюджет всех кампаний, запущенных на определенной платформе**

Функция `calculate_average_budget` должна принимать путь к файлу и название плафтормы и возвращать число - средний бюджет всех кампаний, запущенных на данной платформе, округленный до двух знаков после запятой.

````python

def calculate_average_budget(path, platform):
  lines = []
  with open(path,'r', newline='') as f:
    lines.extend(csv.reader(f, delimiter=','))

  data = [{'platform' : line[4], 'budget' : int(line[5])} for line in lines[1:]]
  l = len(data)

  budget_list = [data[i]['budget'] for i in range(l) if data[i]['platform'] == platform]
  avr_budget = round(mean(budget_list), 2)
  return avr_budget

avr_budget = calculate_average_budget('campaign_data.csv', 'Google')
avr_budget
````

### **Напишите генератор, который будет генерировать даты, в которые НЕ запускались кампании**

Функция `get_missing_campaign_dates` должна принимать путь к файлу и генерировать построчно даты, отсортированные по возрастанию, в которые не запускались кампании, в формате:

```
2022-02-29
2022-03-28
2022-03-29
2022-05-30
2022-06-13
2022-06-27
...
```

````python
from datetime import timedelta

def get_missing_campaign_dates(path):
  lines = []
  with open(path,'r', newline='') as f:
    lines.extend(csv.reader(f, delimiter=','))

  data = [datetime.strptime(line[1], '%Y-%m-%d') for line in lines[1:]]

  for i in range(len(data)-1):
    current_date = data[i] + timedelta(days=1)
    while current_date < data[i+1]:
      print(str(current_date)[:10])
      yield current_date
      current_date += timedelta(days=1)

get_missing_campaign_dates('campaign_data.csv')
````

### **Напишите функцию, которая будет принимать на вход путь к файлу и возвращать сгруппированные данные по городам, содержащие суммарный бюджет и количество кликов**

Функция `group_campaign_data` должна возвращать список словарей, отсортированный по городам, в формате:

```python
[{'Город': 'Нижний Новгород',
  'Количество кликов': 616.0,
  'Суммарный бюджет': 27585},
 {'Город': 'Новосибирск',
  'Количество кликов': 733.0,
  'Суммарный бюджет': 33001},
  ...]
```

````python

def group_campaign_data(path):
  lines = []
  with open(path,'r', newline='') as f:
    lines.extend(csv.reader(f, delimiter=','))

  data = [{'budget' : int(line[5]),
         'Город' : line[6],
         'Количество кликов' : float(line[8]),
         } for line in lines[1:]]
  l = len(data)

  cities = list(set(d['Город'] for d in data))
  cities.sort()

  sum_data = [{'Город': city,
    'Количество кликов': sum(d['Количество кликов'] for d in data if d['Город'] == city),
    'Суммарный бюджет': sum(d['budget'] for d in data if d['Город'] == city)
    } for city in cities]

  return sum_data

group_campaign_data('campaign_data.csv')
````

### **Напишите генератор, который будет генерировать данные о кампаниях, запущенных в указанном городе и имеющих бюджет выше заданного значения**

Функция `campaign_generator` должна принимать путь к файлу, город и бюджет и генерировать словари, отсортированные по возрастанию 'ID Кампании' в формате:

```
{'ID Кампании': '657', 'Тип': 'Контекстная', 'Платформа': 'Вконтакте', 'Доход': '10692'}
{'ID Кампании': '674', 'Тип': 'Нативная интеграция', 'Платформа': 'Яндекс', 'Доход': '20766'}
{'ID Кампании': '761', 'Тип': 'Соцсети', 'Платформа': 'Вконтакте', 'Доход': '15680'}
```

````python
from datetime import timedelta

def campaign_generator(path, City, min_budget):
  lines = []
  with open(path,'r', newline='') as f:
    lines.extend(csv.reader(f, delimiter=','))

  data = [{'ID Кампании': line[0],
         'Тип' : line[3],
         'budget' : int(line[5]),
         'City' : line[6],
         'Платформа' : line[4],
         'Доход' : line[10]
         } for line in lines[1:]]

  data_filt = [d for d in data if d['City'] == City and d['budget'] > min_budget]

  for d in data:
    if d['City'] == City and d['budget'] > min_budget:
      d_filt = {'ID Кампании': d['ID Кампании'],
                'Тип': d['Тип'],
                'Платформа': d['Платформа'],
                'Доход': d['Доход']}
      yield d_filt

list(campaign_generator('campaign_data.csv', 'Москва', 48000))
````

````python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
correct_answer = answers[7]

campaigns = campaign_generator('campaign_data.csv', 'Москва', 48000)
user_answer = []
for campaign in campaigns:
    user_answer.append(campaign)

try:
    assert str(user_answer) == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
````
