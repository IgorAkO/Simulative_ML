### **Посчитайте Retention 15 дня (в процентах) для пользователей, зарегистрированных в январе**

Cохраните результат в переменную `retention_15_day`

**Примечание:** результат округлите до 5 знаков после запятой

```python
import csv
from datetime import datetime, timedelta

registrations = 'registrations.csv'
entries = 'entries.csv'

# Создем словарь d_reg = {id:reg_date}
# и словарь d_reg_01 для reg_date в январе
d_reg = {}
d_reg_01 = {}
with open(registrations,'r', newline='') as f_reg:
  reader = csv.reader(f_reg, delimiter=';')
  next(reader)
  for row in reader:
    # Словарь d_reg
    day = datetime.strptime(row[1], "%Y-%m-%d")
    d_reg[row[0]] = day

    # Словарь d_reg_01
    month = datetime.strptime(row[1], "%Y-%m-%d").month
    if month == 1:
      d_reg_01[row[0]] = day

# Список словарей d_ent = [{'id': , 'date': , 'days': }, {...}, ]
# и d_ent_01_date в январе
set_ent = []
set_ent_01 = []
with open(entries,'r', newline='') as f_ent:
  reader = csv.reader(f_ent, delimiter=';')
  next(reader)
  for row in reader:
    # Словарь d_ent
    date = datetime.strptime(row[1], "%Y-%m-%d")
    days = (date - d_reg[row[0]]).days
    d_ent = {'id' : row[0], 'date': date, 'days' : days}
    if d_ent not in set_ent:
      set_ent.append(d_ent)

      # Словарь d_ent_01
      month = datetime.strptime(row[1], "%Y-%m-%d").month
      if month == 1:
        set_ent_01.append(d_ent)

# Создаем сkоварь для Retention d_ret = {id:day}
def retention(N_days, d_reg, set_ent):
  d_ret = {}
  for ent in set_ent:
    if (ent['id'] not in d_ret.keys()) and (ent['days'] == N_days):
      d_ret[ent['id']] = ent['days']

  retention_N_day = round(100 * len(d_ret) / len(d_reg), 5)
  return retention_N_day

retention_15_day = retention(15, d_reg_01, set_ent_01)
retention_15_day
```

```python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
# Открываем файл с правильными ответами
with open('metrics.txt', 'r') as f:
    answers = f.read().split('\n')

correct_answer = float(answers[0])

try:
    assert retention_15_day == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```

### **Посчитайте Rolling-retention 30 дня (в процентах) для пользователей из той же когорты**

Сохраните результат в переменную `rolling_retention`

**Примечание:** результат округлите до 5 знаков после запятой

```python
# Создаем словарь для Rollling Retention d_rol_ret = {id:day}
def rolling_retention(N_days, d_reg, set_ent):
  d_rol_ret = {}
  for ent in set_ent:
    if (ent['id'] not in d_ret.keys()) and (ent['days'] >= N_days):
      d_rol_ret[ent['id']] = ent['days']

  rolling_retention_N_day = round(100 * len(d_rol_ret) / len(d_reg), 5)
  return rolling_retention_N_day

rolling_retention = retention(30, d_reg_01, set_ent_01)
rolling_retention
```

```python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
correct_answer = float(answers[1])

try:
    assert rolling_retention == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```

### **Посчитайте Lifetime по всем пользователям, посчитанный как интеграл от n-day retention**

Сохраните результат в переменную `lifetime`

**Примечание:** результат округлите до 5 знаков после запятой

```python
def Lifetime(d_reg, set_ent):
  lifetime = 0
  ret = 1
  i = 0
  while ret > 0.00001:
    ret = retention(i, d_reg, set_ent)
    lifetime += ret / 100
    #print(f'{i}: {ret}, {lifetime}')
    i += 1
  return lifetime

lifetime = round(Lifetime(d_reg, set_ent), 5)
lifetime
```

### **Посчитайте Churn rate 29 дня (в долях), посчитанный по всем пользователям**

Сохраните результат в переменную `churn_29`.  
Если будете считать CR от Retention, ведите расчет от Rolling Retention, таким образом, мы получим всех, кто не заходил в 29 день и после. Ведя расчет от обычного Retention, наоборот - получим только CR в 29 день.


```python
def churn_rate(N_days, d_reg, set_ent):
  d_churn = list(d_reg.keys())
  for ent in set_ent:
    if (ent['id'] in d_churn) and (ent['days'] >= N_days):
      d_churn.remove(ent['id'])

  churn_rate_N = round(len(d_churn) / len(d_reg), 5)
  return  churn_rate_N

churn_29 = churn_rate(29, d_reg, set_ent)
churn_29
```

### **Посчитайте Mau, Wau, Dau за последний месяц/неделю/день записей**

Сохраните результат в переменные `dec_mau`, `dec_wau`, `dec_dau` соответственно

**Примечание:** последний месяц записей - декабрь. Поэтому `mau` рассчитываем для декабря (2021 года), для `wau` берем последнюю неделю - с 25 по 31 декабря, и для `dau` соответственно последний день - 31 декабря.

```python
def mau(month, set_ent):
  user_month = []
  for ent in set_ent:
    m_ent = ent['date'].month
    if m_ent == month and ent['id'] not in user_month:
      user_month.append(ent['id'])
  return len(user_month)

dec_mau = mau(12, set_ent)
dec_mau
```
